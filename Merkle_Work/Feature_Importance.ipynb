```python
import pandas as pd
adv_stats = pd.read_csv("../data/adv_team_2018.csv")
adv_stats.head()
```

```python
adv_stats.fillna(0,inplace=True)
adv_stats = adv_stats.drop(["Team","Opp.Team","gameid","PTS","Opp.PTS","MP","X2P","X2PA","X3P","X3PA",
                           "Opp.2P","Opp.2PA","Opp.3P","Opp.3PA","FG","FGA","Opp.FG","Opp.FGA",
                           "FT","FTA","Opp.FT","Opp.FTA","FT.","Opp.FT.",
                            "FG.","Opp.FG.","X2P.","X3P.","Opp.2P.","Opp.3P.",
                           "PF","Opp.PF","Opp.ORB","ORB","Opp.TRB","TRB",
                           "AST","TOV","Opp.AST","Opp.TOV","ppp","Opp.ppp"],axis=1)
adv_stats.head()
```

```python
wins = adv_stats[["Win."]]
adv_stats = adv_stats.drop(["Win."],axis=1)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(adv_stats,wins,test_size = 0.30, random_state=22)
```

```python
from sklearn.ensemble import ExtraTreesClassifier
from sklearn import ensemble
```

```python
extc = ExtraTreesClassifier(n_estimators=1000,criterion='gini',min_samples_split= 5,
                            min_samples_leaf= 5)   
```

```python
extc.fit(X_train,y_train)
```

```python
import numpy as np
importances = extc.feature_importances_
indices = np.argsort(importances)[::-1]
```

```python
print("Feature ranking:")

for f in range(X_train.shape[1]):
    print("%d. %s (%.3f) (%.3f)" % (f + 1, X_train.columns[indices[f]], importances[indices[f]]*100,
                                    np.sum(importances[indices[0:f+1]]*100)))
```

```python
y_pred = extc.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_pred)
```
